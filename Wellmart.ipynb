```python
df.groupby('Holiday_Flag')['Weekly_Sales'].agg(['sum','mean','std'])
```

```python
Holiday weeks generate higher average sales but also show higher volatility compared to non-holiday weeks, 
indicating increased but less predictable consumer demand during holidays.
```

```python
store_summary = df.groupby('Store')['Weekly_Sales'] \
    .agg(['sum','mean','std'])

store_summary.sort_values('mean', ascending=False).head(5)
```

```python
Store 20, 4, and 14 are the top-performing stores based on average weekly sales. 
    However, they also show higher volatility, indicating demand fluctuations. 
    Smaller stores such as Store 33 and 30 are more stable but generate significantly lower revenue.
```

```python
store_summary = df.groupby('Store')['Weekly_Sales'] \
    .agg(['sum','mean','std'])

store_summary.sort_values('mean', ascending=True).head(5)
```

```python
Store 33 has the lowest average weekly sales of around 260 thousand, indicating weaker performance.
```

```python
df.groupby('Month')['Weekly_Sales'].mean().sort_values(ascending=False)
```

```python
I performed month-wise sales analysis and observed strong seasonality. December showed the highest average sales due to holiday demand, 
while January had the lowest performance, indicating post-holiday spending decline.
```

```python
df['Quarter'] = df['Date'].dt.quarter
df.groupby('Quarter')['Weekly_Sales'].mean()
```

```python
Quarter-wise analysis revealed strong seasonality, with Q4 generating the highest average sales due to festive demand, 
while Q1 showed relatively lower performance following the holiday season.
```

```python
year_sales = df.groupby('Year')['Weekly_Sales'].mean()
year_sales.pct_change()*100
```

```python
Year-over-year analysis showed a slight decline of approximately 1% in 2011 and 2012, 
indicating relatively stable but slightly decreasing sales performance over time.
```
